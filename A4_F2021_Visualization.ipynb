```python
# create 3 new columns:
# 'Total_Minutes': sum of total day, eve, night, and international minutes
# 'Total_Charges' sum of total day, eve, night, and international charges
# 'Total_Calls' sum of total day, eve, night, and international calls

orange_df['Total_Minutes'] = orange_df['Total day minutes'] + \
    orange_df['Total eve minutes'] + orange_df['Total night minutes'] + \
    orange_df['Total intl minutes']

orange_df['Total_Charges'] = orange_df['Total day charge'] + \
    orange_df['Total eve charge'] + orange_df['Total night charge'] + \
    orange_df['Total intl charge']

orange_df['Total_Calls'] = orange_df['Total day calls'] + \
    orange_df['Total eve calls'] + orange_df['Total night calls'] + \
    orange_df['Total intl calls']

orange_df
```

```python
# create 2 additional columns
# 'Avg_Rate': the average charge per minute (use columns created above to help you make the new column)
# 'Cost_Per_Call': the average charge per call (use columns created above to help you make the new column)

orange_df['Avg_Rate'] = orange_df['Total_Charges'] / orange_df['Total_Minutes']
orange_df['Cost_Per_Call'] = orange_df['Total_Charges'] / orange_df['Total_Calls']
orange_df



```

```python
# Create the first scatter plot
# Use the x-axis for number of calls
# Use the y-axis for the cost per call
# Label both axes
# play with plotting parameters (size, alpha, figure size, etc.) to make the plot more appealing
# make the figure size 10 x 8

ax1 = orange_df.plot.scatter(x= 'Total_Calls', y= 'Cost_Per_Call', label= 'Total Calls v. CPC')
import matplotlib.pyplot as plt
figure = plt.figure(figsize=(10, 8))

figure



```
